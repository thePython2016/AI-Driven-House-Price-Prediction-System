```python
import pandas as pd
import numpy as np
import seaborn as sns
import xgboost
from xgboost import XGBClassifier
from sklearn.metrics import precision_score,recall_score,accuracy_score, confusion_matrix,classification_report
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
```

```python
studyHours=[2,4,3,5,9,5,2]
method=["Online","Offline","Hybrid","Online","Offline","Online","Hybrid"]
performance=["Poor","Poor","Good","Poor","Good","Worst","Good"]

df=pd.DataFrame({"Hours":studyHours,"Method":method,"Performance":performance})
```

```python
df
```

```python
x=df.drop(columns=['Performance'])
y=df['Performance']
```

```python
y
```

```python
ordinal=OrdinalEncoder(categories=[['Offline','Online','Hybrid']])
# # ordinal
x['Method']=ordinal.fit_transform(x[['Method']])
```

```python
x
```

```python
label=LabelEncoder()
y=pd.DataFrame(y,columns=['Performance'])
y=label.fit_transform((y).values.ravel())
```

```python
y
```

```python
model=XGBClassifier()
model.fit(x,y)
```

```python
predicted=model.predict(x)
predicted
```

```python
score=precision_score(y,predicted,average="weighted")
```

```python
report=classification_report(y,predicted,target_names=label.classes_.astype(str),output_dict=False)
print(report)
```

```python
classes=model.classes_
```

```python
target_names = ['Dropout', 'Enrolled', 'Graduate']

report = classification_report(
    y,
    predicted,
    target_names=label.classes_,  # ✅ manually defined
    zero_division=0
)
print(report)
```
